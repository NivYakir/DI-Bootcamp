# Create an Excel workbook using openpyxl with a sheet named “Grades”

```python
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill
from openpyxl.utils import get_column_letter

# Create a new workbook
wb = Workbook()
ws = wb.active
ws.title = "Grades"

# Data structure
data = {
    "Joe": {"math": 65, "science": 78, "english": 98, "gym": 89},
    "Bill": {"math": 55, "science": 72, "english": 87, "gym": 95},
    "Tim": {"math": 100, "science": 45, "english": 75, "gym": 92},
    "Sally": {"math": 30, "science": 25, "english": 45, "gym": 100},
    "Jane": {"math": 100, "science": 100, "english": 100, "gym": 60},
}

# 1️⃣ Write header row
headers = ["Student", "Math", "Science", "English", "Gym"]
ws.append(headers)

# 2️⃣ Write data rows
for student, grades in data.items():
    row = [student] + list(grades.values())
    ws.append(row)

# 3️⃣ Add average formulas
avg_row = ws.max_row + 1  # Row Number Where Avg's will be calculated
ws.cell(row=avg_row, column=1, value="Average") # Cell(avg_row, 1) = 'Average'

# Loop through each cell (column by column)
for col in range(2, ws.max_column + 1):
    col_letter = get_column_letter(col)
    ws.cell(row=avg_row, column=col, value=f"=AVERAGE({col_letter}2:{col_letter}{ws.max_row - 1})")

# 4️⃣ Apply header formatting
header_fill = PatternFill(start_color="FFC000", end_color="FFC000", fill_type="solid")
for cell in ws[1]:
    cell.font = Font(bold=True)
    cell.fill = header_fill

# 5️⃣ Apply bold formatting to averages
for cell in ws[avg_row]:
    cell.font = Font(bold=True)

# 6️⃣ Save workbook
wb.save("example_workbook.xlsx")
```
