```python
import pandas as pd
import numpy as np
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.styles import PatternFill
from openpyxl.utils import get_column_letter
from openpyxl.utils.dataframe import dataframe_to_rows
```

# Create an Excel workbook using openpyxl with a sheet named “Grades”

```python
from openpyxl import Workbook

# Create a new workbook
wb = Workbook()

# Get the active worksheet and rename it
ws1 = wb.active
ws1.title = "Grades"

# Populate the sheet with the following data structure:
data = {
    "Joe": {
        "math": 65,
        "science": 78,
        "english": 98,
        "gym": 89
    },
    "Bill": {
        "math": 55,
        "science": 72,
        "english": 87,
        "gym": 95
    },
    "Tim": {
        "math": 100,
        "science": 45,
        "english": 75,
        "gym": 92
    },
    "Sally": {
        "math": 30,
        "science": 25,
        "english": 45,
        "gym": 100
    },
    "Jane": {
        "math": 100,
        "science": 100,
        "english": 100,
        "gym": 60
    }
}

# First Convert Data to Pandas Dataframe
df = pd.DataFrame(data).T

# Insert DataFrame into Worksheet
for row in dataframe_to_rows(df, index=True, header=True):
    if all(cell is None for cell in row):
        continue  # skip empty rows
    ws1.append(row)

# Set Cell A1 = 'Student'
ws1['A1'] = 'student'

# Calculate Average Grade for Each Column
# Determine number of rows and columns
max_row = ws1.max_row
max_col = ws1.max_column

# Write "Average" label in first column after the last row
ws1.cell(row=max_row + 1, column=1, value="Average")

# Loop through each column and insert AVERAGE formula
for col in range(2, max_col + 1):
    col_letter = get_column_letter(col)
    avg_formula = f"=AVERAGE({col_letter}2:{col_letter}{max_row})"
    ws1.cell(row=max_row + 1, column=col, value=avg_formula)

# Apply bold and colored formatting to the header row.
header_fill = PatternFill(start_color="FFC000", end_color="FFC000", fill_type="solid")
for cell in ws1[1]:
    cell.font = Font(bold=True)
    cell.fill = header_fill

# Apply bold formatting to the formula row.
for cell in ws1[ws1.max_row]:
    cell.font = Font(bold=True)

# Save the workbook to a file
wb.save('example_workbook.xlsx')
```
