```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from urllib.request import urlopen
import requests
```

```python
url = 'https://github.com/topics'
response = requests.get(url)

# Status Check
print("Status Code =", response.status_code)

# Print First 100 Chars of Response
print(response.text[:100])

# Save HTML Content to a File
with open("webpage.html", "w", encoding="utf-8") as file:
    file.write(response.text)
```

```python
# Open and read the saved HTML file
with open('webpage.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Use BeautifulSoup to parse the saved HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Identify two distinct pieces of information on the webpage to extract (Topic Titles and Their Descriptions)
# Create a List of all the Topic Elements
topics = soup.find(class_='col-lg-9 position-relative pr-lg-5 mb-6 mr-lg-5').find_all(class_='py-4 border-bottom d-flex flex-justify-between')

# Extract Title Text from each Element
titles = [t.find(class_='f3 lh-condensed mb-0 mt-1 Link--primary').get_text() for t in topics]

# Extract Description Text from each Element
descriptions = [d.find(class_='f5 color-fg-muted mb-0 mt-1').get_text(strip=True) for d in topics]

# Close File
file.close()

# Check
print(f"Titles: {len(titles)} Items")
for i, t in enumerate(titles):
    print(f"{i+1}: {t}")
    
print(f"\nDescriptions: {len(descriptions)} Items")
for i, d in enumerate(descriptions):
    print(f"{i+1}: {d}")
```

```python
# Create a Python dictionary to structure the extracted data
my_dict = {
    'title': titles,
    'description': descriptions
}

# Convert this dictionary into a pandas DataFrame
df = pd.DataFrame(my_dict)

# Print the DataFrame to confirm its structure and contents.
df
```
