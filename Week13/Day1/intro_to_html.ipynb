```python
html = """
<html>
     <head>
         <title>2020 Oscar nominees and winners</title>
     </head>
     <body>
         <h1>Oscars 2020</h1>
         <p>The biggest cinema awards ceremony took place in February. This year's highlights were:</p>
         <ul class="highlights">
             <li>Joker</li>
             <li>1917</li>
             <li>Once Upon a Time in Hollywood</li>
             <li>The Irishman</li>
         </ul>
         <p>See the winners of 3 categories below.</p>
         <div>
             <div class="best-film-category">
                 <h2>Category: best film</h2>
                 <ul>
                     <li>Ford vs Ferrari</li>
                     <li>The Irishman</li>
                     <li>JoJo Rabbit</li>
                     <li>Joker</li>
                     <li>Little Women</li>
                     <li>Marriage Story</li>
                     <li>1917</li>
                     <li>Once Upon a Time in Hollywood</li>
                     <li class="winner">Parasite <strong>[WINNER]</strong></li>
                 </ul>
             </div>
             <br>
             <div class="best-actor-category">
                 <h2>Category: best actor</h2>
                 <ul>
                     <li>Antonio Banderas - Pain and Glory</li>
                     <li>Leonardo DiCaprio - Once Upon a Time In... Hollywood</li>
                     <li>Adam Driver - Marriage Story</li>
                     <li class="winner">Joaquin Phoenix - Joker <strong>[WINNER]</strong></li>
                     <li>Jonathan Price - The Two Popes</li>
                 </ul>
             </div>
             <br>
             <div class="best-actress-category">
                 <h2>Category: best actress</h2>
                 <ul>
                     <li>Cythia Erivo - Harriet</li>
                     <li>Scarlett Johansson - Marriage Story</li>
                     <li>Saoirse Ronan - Little Women</li>
                     <li>Charlize Theron - The Scandal</li>
                     <li class="winner">Renée Zellweger - Judy: Far Over the Rainbow <strong>[WINNER]</strong></li>
                 </ul>
             </div>
         </div>
     </body>
</html>
"""
```

```python
from bs4 import BeautifulSoup
# Instantiate a BeautifulSoup Object
soup = BeautifulSoup(html, 'html.parser') 
print(type(soup))

# To format HTML/XML content in a way that's easier to read, we use prettify()
print(soup.prettify())
```

#### Collecting With Attributes

```python
print(soup.find_all("div", {"class": "best-actress-category"}))

# OR

print(soup.find_all(class_="best-actress-category"))

# What is the length of these lists (Should be 1)
print(len(soup.find_all(class_="best-actress-category")))
```

```python
# Retrieving the text itself
for item in soup.find_all("div", {"class": "best-actress-category"}):
    print(item.get_text())

# In order to only retrieve the Actress Names, you need to go a further step
div = soup.find("div", {"class" : "best-actress-category"})
actresses = div.find_all("li")

for a in actresses:
    print(a.get_text())
```

# SUMMARY AND EXPLANATION OF FIND ALL AND SOUP OBJECTS

```python
# find_all() ALWAYS returns a list
div_list = soup.find_all("div", {"class": "best-actress-category"}) # This returns a list with 1 item, the entire <div> that has the 'actresses' name since there is only 1
print(len(div_list)) # 1

# all_list is a BeatifulSoup Tag Object, representing the entire <div> block
# Inside that single list item (the <div> tag), you have nested tags:
# one <h2>
# one <ul>
# several <li> elements
# one <strong> tag nested within the winner <li>
# But these nested tags are not separate items in the list returned by find_all() —
# they are children of the one <div> element.

div = div_list[0]
print(div.name)
print(len(div.find_all("li")))  # prints 5 (the number of actresses)
```

### Navigating with BeautifulSoup
#### Let's find the WINNERS in each category

```python
# First we select the <div> best-film-category
best_film_winner = soup.find('div', class_='best-film-category').find('li', class_='winner') # Finds the specific class with that name

# Print the result text
print(f"Best Film Winner: {best_film_winner.get_text()}")

# use .parent to navigate UP the tree
winner_parent = best_film_winner.parent
print(f"Winner's Parent Tag: {winner_parent.name}")
category_div = winner_parent.parent # Get the parent <div> of the winner (the category container)
print(f"Parent's Parent Tage: {category_div.name}")

# Using .next_sibling and .previous_sibling
next_category = category_div.find_next_sibling('div')
previous_category = category_div.find_previous_sibling('div')
print(f"Next Category: {next_category.h2.get_text() if next_category else 'None'}")
print(f"Previous Category: {previous_category.h2.get_text() if previous_category else 'None'}")

# Navigating to the Next Element
# BeautifulSoup treats tags, strings, and comments as elements (nodes) in the HTML tree.
next_element = best_film_winner.find_next()
print(f"Next Element after Winner: {next_element}")

# If you want to skip text and only get the next tag element, you can use:
next_tag_element = best_film_winner.find_next_sibling()
print(f"Next Tag Element: {next_tag_element}")
```
