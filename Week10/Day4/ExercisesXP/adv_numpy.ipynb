# Exercise 6 : Basic Hypothesis Testing

```python
# Create a sample dataset to test the effectiveness of a new training program on employee productivity
# Set random seed for reproducibility


# Productivity scores before and after training
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

print("=" * 60)
print("HYPOTHESIS TEST: Training Program Effectiveness")
print("=" * 60)

# Formulate hypotheses
print("\nHypotheses:")
print("H₀ (Null): The training program has no effect (μ_difference = 0)")
print("H₁ (Alternative): The training program increases productivity (μ_difference > 0)")

# Calculate the difference
productivity_change = productivity_after - productivity_before

# Descriptive statistics
print(f"\nDescriptive Statistics:")
print(f"Mean productivity before:  {productivity_before.mean():.2f}")
print(f"Mean productivity after:   {productivity_after.mean():.2f}")
print(f"Mean change:               {productivity_change.mean():.2f}")
print(f"Std deviation of change:   {productivity_change.std():.2f}")

# Perform PAIRED t-test (one-tailed)
t_statistic, p_value_two_tailed = stats.ttest_rel(productivity_before, productivity_after)
p_value_one_tailed = p_value_two_tailed / 2

# Define significance level
alpha = 0.05

print(f"\nTest Results:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value (one-tailed): {p_value_one_tailed:.4f}")
print(f"Significance level (α): {alpha}")

# Decision
print(f"\nDecision:")
if p_value_one_tailed < alpha and t_statistic < 0:
    print(f"Reject H₀: Strong evidence that training increased productivity")
    print(f"  (p = {p_value_one_tailed:.4f} < {alpha})")
else:
    print(f"Fail to reject H₀: Insufficient evidence of improvement")
    print(f"  (p = {p_value_one_tailed:.4f} ≥ {alpha})")
```

# Exercise 8 : Time Series Data Manipulation

```python
# Generate time series data for the year 2023
dates = np.arange('2023-01', '2024-01', dtype='datetime64[D]')

# Demonstrate Slicing for the following intervals:
ts = pd.Series(dates, index=dates)

# January - March
winter = ts['2023-01':'2023-03']
print(winter)

# April to June
spring = ts['2023-04' : '2023-06']
print(spring)

# July to September
summer = ts['2023-07': '2023-09']
print(summer)

# October to December
fall = ts['2023-10' : '2023-12']
print(fall)
```
