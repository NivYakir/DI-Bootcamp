# Exercise 1: Multivariate Normal Distribution

```python
from scipy.stats import norm
from scipy.stats import multivariate_normal

# Create Univariate Distribution
univariate_data = norm.rvs(size=1000)

# Create MultiVariate Distribution
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance, independent variables
multivariate_data = multivariate_normal.rvs(mean=mean, cov=cov, size=1000)
```

```python
# --- Create Plots ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left plot: Univariate
sns.histplot(univariate_data, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title("Univariate Normal Distribution")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

# Right plot: Multivariate
axes[1].scatter(multivariate_data[:, 0], multivariate_data[:, 1], alpha=0.6, color='red')
axes[1].set_title("Multivariate Normal Distribution (2D)")
axes[1].set_xlabel("Variable 1")
axes[1].set_ylabel("Variable 2")
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```
