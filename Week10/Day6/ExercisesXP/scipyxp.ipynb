# Exercise 2: Descriptive Statistics

```python
# Task: Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
data = [12, 15, 13, 12, 18, 20, 22, 21]

description = scipy.stats.describe(data)

print(f"Number of Observations: {description.nobs}")
print(f"Min and Max: {description.minmax}")
print(f"Mean: {description.mean:.2f}")
print(f"Variance: {description.variance:.2f}")
print(f"Skewness: {description.skewness:.2f}")
print(f"Kurtosis: {description.kurtosis:.2f}")
```

# Exercise 3: Understanding Distributions

```python
import matplotlib.pyplot as plt
from scipy import stats

# Task: Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.
mu, sig = 50, 10

# Create Distribution using Scipy Stats
norm_dist = stats.norm(mu, sig)
print("PDF at x=0.5:", norm_dist.pdf(0.5))

# Plot Using Matplotlib
x_gaussian = np.linspace(0, 100, 1000) # determines the range of the x-axis
pdf_gaussian = norm_dist.pdf(x_gaussian) # creates a probability density function from the distribution created earlier
plt.figure(figsize=(6, 4))
plt.plot(x_gaussian, pdf_gaussian)
plt.title('Gaussian Distribution\n$\mu=50$, $\sigma=10$')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

# Exercise 4: T-Test Application

```python
# Task: Perform a T-test on two sets of randomly generated data.
np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Two-Sample Independent T-Test (comparing the means of 2 independent samples)
t_stat, p_val = stats.ttest_ind(data1, data2)
print('T-Statistic:', t_stat)
print('P-Value:', p_val)

if p_val < 0.05:
    print("There is a statistically significant difference between the two groups")
else:
    print("There is no statistically siginificant difference between the two groups")
```

# Exercise 5: Linear Regression Analysis

```python
# Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), 
# Use linear regression to predict the price of a house given its size.
data = pd.DataFrame({
    'house_size': [50, 70, 80, 100, 120],
    'house_price': [150000, 200000, 210000, 250000, 280000]
})

# What is the slope/intercept of the regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data['house_size'], data['house_price'])
print(f"Slope: {slope:.3f}\nIntercept: {intercept:.3f}")

# Predict the price of a house that is 90 square meters
# use Y = mX + b (Where X = 90, and m/b were calculated earlier)
result = (slope * 90) + intercept
print(f"Price of House Size 90: {result:.3f}")
```

# Exercise 6: Understanding ANOVA

```python
# Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. 
# The growth in centimeters is recorded.
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Perform an ANOVA test on the given data. What are the F-value and P-value?
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f"F-value: {f_value:.2f}")
print(f"P-value: {p_value:.5f}")

# Based on the P-value, do the fertilizers have significantly different effects on plant growth?
if p_value < 0.05:
    print("Reject the Null Hypothesis. The Fertilizers DO have significantly different effects on plant growth")

# Explain what would happen if the P-value were greater than 0.05
print("We accept the Null Hypothesis. Meaning the difference in growth between the fertilizers is NOT statistically significant.")
```

# Exercise 7: Probability Distributions (Optional)

```python
# Work with a binomial distribution: calculate probabilities for different numbers of successes
# Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.
n, p = 10, 0.5  # number of trials, probability of each trial
binom_dist = stats.binom(n, p)
print("PMF at 5 successes:", binom_dist.pmf(5))

# Plotting Binomial Distribution
x_binom = np.arange(0, n+1)
pmf_binom = binom_dist.pmf(x_binom)
plt.figure(figsize=(6, 4))
plt.bar(x_binom, pmf_binom)
plt.title('Binomial Distribution of 10 Coin Flips\n$n=10$, $p=0.5$')
plt.xlabel('Number of Successes')
plt.ylabel('PMF')
plt.show()
```

# Exercise 8: Correlation Coefficients (Optional)

```python
from scipy.stats import pearsonr, spearmanr

# Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

# Pearson Correlation
corr, _ = pearsonr(data['age'], data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')

# Spearman Correlation
spcorr, _ = spearmanr(data['age'], data['income'])
print(f'Spearman Rank Correlation: {spcorr:.2f}')
```
