Hypothesis Testing
What is Hypothesis Testing?

Hypothesis testing is a statistical method used to determine the likelihood of a hypothesis being true based on sample data. It involves making an initial assumption (the null hypothesis) and testing this assumption to infer about a population.

image.png

image.png

```python
import numpy as np
from scipy import stats

# Sample engagement times (in minutes)
sample_data = np.array([17, 18, 21, 20, 16, 19, 22, 20, 18, 21,
                        17, 20, 19, 18, 21, 20, 19, 22, 18, 19,
                        20, 21, 20, 19, 21, 20, 22, 19, 18, 21])

print("Sample Data Mean:", np.mean(sample_data))

# Step 1: Test against H₀: μ = 15
t_statistic, p_value = stats.ttest_1samp(sample_data, 15)

# Step 2: Define significance level
alpha = 0.05

# Step 3: Decision
print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: Engagement time has increased.")
else:
    print("Fail to reject the null hypothesis: Not enough evidence of increase.")
```

How spread out your data is (Variance)

Whether the data is symmetrical or skewed (Skewness)

Whether it has heavy or light tails (Kurtosis)

Skewness = 0 → perfect symmetry (normal distribution)

Positive skew → tail on the right (more high values)

Negative skew → tail on the left (more low values)

```python
from scipy.stats import skew

data = np.array([2, 4, 6, 8, 50])  # Extreme high value
skewness = skew(data)
print("Skewness:", skewness)
```

Kurtosis measures the "tailedness" of the distribution (how much data is in the tails vs. the center).

Kurtosis ≈ 3 → normal distribution (mesokurtic)

Kurtosis > 3 → heavy tails (leptokurtic) → more outliers

Kurtosis < 3 → light tails (platykurtic) → fewer outliers

```python
from scipy.stats import kurtosis

data = np.array([3, 5, 7, 9, 100])
kurt = kurtosis(data)  # Excess kurtosis
print("Excess Kurtosis:", kurt)
```

Time Series Use Case with Pandas + NumPy


While NumPy handles low-level date math efficiently, Pandas provides higher-level time series tools.

```python
import numpy as np
import pandas as pd

# Step 1: Generate a time series with 30 random values and daily date index
ts = pd.Series(np.random.randn(30), index=pd.date_range('2023-01-01', periods=30))
print(ts)
# Step 2: Apply a 5-day rolling window and calculate the mean for each window
rolling_avg = ts.rolling(window=5).mean() # Can Change Aggregation Type to SUM/Median/etc. as well

# Step 3: Print the result
print("\n5-Day Rolling Average:\n", rolling_avg)
```

```python
import pandas as pd

# Create a time series with 100 random values and date index
ts_data = pd.Series(np.random.randn(100), index=pd.date_range('2023-01-01', periods=100))
print(ts_data)
# Rolling mean over a 5-day window
rolling_mean = ts_data.rolling(window=5).mean()
print("Rolling Mean:\n", rolling_mean)

```
