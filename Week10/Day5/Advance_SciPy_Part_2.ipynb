```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson
n = 10    # number of trials
p = 0.5   # probability of success
x = np.arange(0, n+1)
pmf_binom = binom.pmf(x, n, p)
plt.bar(x, pmf_binom, color='teal')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass')
plt.grid(True)
```

```python
# Example: Binomial Distribution
n, p = 10, 0.5  # number of trials, probability of each trial
binom_dist = binom(n, p)
print("PMF at 5 successes:", binom_dist.pmf(5))
```

```python


n, p = 10, 0.5
binom_dist = binom(n, p)

# Evaluate PMF
x = np.arange(0, n+1)
pmf = binom_dist.pmf(x)

# Plot
plt.figure(figsize=(6, 4))
plt.bar(x, pmf)
plt.title('Binomial Distribution\nn=10, p=0.5')
plt.xlabel('Number of Successes')
plt.ylabel('PMF')
plt.grid(True)
plt.show()
```

```python
lambda_ = 3  # average number of events
x = np.arange(0, 9)
pmf_poisson = poisson.pmf(x, mu=lambda_)
plt.bar(x, pmf_poisson, color='teal')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability Mass')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Poisson Example
lambda_ = 3  # average number of calls per hour
poisson_dist = poisson(lambda_)
print("PMF for 3 events:", poisson_dist.pmf(3))  #So theres about a X% chance of receiving exactly 3 calls in the next hour.
```

```python
from scipy import stats
import numpy as np

# Simulate two groups with the same mean (for now)
group1 = stats.norm.rvs(loc=5, scale=10, size=100)
group2 = stats.norm.rvs(loc=5, scale=10, size=100)

# Perform independent two-sample T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Output
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt


# Step 3: Load the dataset
df = pd.read_csv('Sample-Data-Plant-Growth.csv')

# Preview the first few rows
print("First few rows of the dataset:")
print(df.head())

# Use a T-test to determine if different fertilizers significantly affect plant growth.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize plant growth by group
plt.figure(figsize=(8, 5))
sns.boxplot(x="Group", y="Weight", data=df)
plt.title("Plant Growth by Fertilizer Group")
plt.ylabel("Plant Weight")
plt.xlabel("Group")
plt.show()
```

```python
from scipy.stats import ttest_ind

# Separate the groups
control = df[df['Group'] == 'control']['Weight']
treatment1 = df[df['Group'] == 'treatment1']['Weight']
treatment2 = df[df['Group'] == 'treatment2']['Weight']

# Run T-tests: control vs trt1, control vs trt2, trt1 vs trt2
# Control vs Trt1
t_1, p_1 = ttest_ind(control, treatment1)
print("Control vs Treatment1")
print("T-statistic:", t_1)
print("P-value:", p_1)

# Control vs Trt2
t_2, p_2 = ttest_ind(control, treatment2)
print("\nControl vs Treatment2")
print("T-statistic:", t_2)
print("P-value:", p_2)

# Trt1 vs Trt2
t_3, p_3 = ttest_ind(control, treatment2)
print("\nTreatment1 vs Treatment2")
print("T-statistic:", t_3)
print("P-value:", p_3)
```

```python
from scipy import stats

# Sample data from three different groups
group1 = [20, 21, 19, 22, 24]
group2 = [28, 30, 27, 26, 29]
group3 = [33, 35, 34, 31, 32]

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(group1, group2, group3)
print("F-value:", f_value)
print("P-value:", p_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Define the dataset
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

# Step 2: Apply linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the regression results
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)

# Step 3: Plot the data points and regression line
plt.figure(figsize=(7, 4))
plt.scatter(x, y, color='blue', label='Data points')  # Original data
plt.plot(x, slope * x + intercept, color='red', label='Regression line')  # Fitted line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

income = [2000, 2500, 3000, 3500, 4000]
spending = [1500, 1800, 2100, 2500, 2800]

corr, _ = pearsonr(income, spending)
print(f"Pearson Correlation: {corr:.2f}")

cov = np.cov(income, spending)[0, 1]
print(f"Covariance: {cov:.2f}")
```

```python
import numpy as np

temperature = [20, 25, 30, 35, 40]
ice_cream_sales = [100, 200, 300, 400, 500]

cov = np.cov(temperature, ice_cream_sales)[0, 1]
corr, _ = pearsonr(temperature, ice_cream_sales)
print(f"Covariance: {cov:.2f}")
print(f"Pearson Correlation: {corr:}")
```
