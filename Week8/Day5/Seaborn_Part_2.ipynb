```python
import matplotlib.pyplot as plt

# Basic histogram using Matplotlib
plt.hist(data['total_bill'], edgecolor='black')  # default bins=10
plt.title('Histogram of Total Bill (Matplotlib)')
plt.xlabel('Total Bill')
plt.ylabel('Count')
#plt.grid(True)
plt.show()

# Customized histogram using Matplotlib
plt.hist(data['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Bill with Custom Bins and Color')
plt.xlabel('Total Bill')
plt.ylabel('Count')
#plt.grid(True)
plt.show()
```

Kernel Density Estimation (KDE) is a way to estimate the probability density function (PDF) of a continuous variable. In simpler terms:

It shows you where the data is concentrated  like a smoothed-out version of a histogram.

Instead of showing counts in bins (like a histogram), KDE draws a smooth curve that represents the shape of the data distribution.

```python
sns.kdeplot(data['total_bill'], shade=True)
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], bw_adjust=0.5, color='red', shade=True)
plt.title('Customized KDE of Total Bill')
plt.show()
```

```python
sns.distplot(data['total_bill'], color='green')
plt.title('Univariate Distribution of Total Bill')
plt.show()

sns.histplot(data['total_bill'], kde=True)
plt.title('Customized KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter')
plt.title('Bivariate Distribution of Total Bill and Tips', pad=70)
plt.show()
```

import seaborn as sns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
titanic = sns.load_dataset('titanic')

# Creating a bar plot
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='class', y='survived', hue='sex', data=titanic, palette='Set2') #, ci='sd'
plt.title('Survival Rate by Class and Gender', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# Creating a box plot
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='class', y='age', data=titanic, palette='Pastel1', width=0.6)
sns.swarmplot(x='class', y='age', data=titanic, size=3, color='k', alpha=0.6)
plt.title('Age Distribution by Class with Individual Data Points', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

```python
# Creating a violin plot
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True,
               palette='coolwarm', inner='quartile', bw=0.8)
plt.title('Age Distribution by Class and Gender', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
data = sns.load_dataset('tips')

# Create scatter plot with regression line
sns.lmplot(x='total_bill', y='tip', data=data)

# Add labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Create a joint plot with regression line
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')

# Add a title
plt.subplots_adjust(top=0.9)
plt.suptitle('Joint Plot of Total Bill and Tip')
plt.show()
```

 What This Plot Shows:
 Main (Center) Plot: Scatter with Regression Line
Each point = one customers total_bill and tip.

The regression line shows the general positive relationship:
 As the total bill increases, tips also tend to increase.

The shaded area around the line is the confidence interval (default is 95%).

 Top Plot: Histogram + KDE of total_bill
Most bills are clustered between 10 and 20 dollars.

The distribution is right-skewed  fewer high bills, with a long tail on the right.

 Right Plot: Histogram + KDE of tip
Most tips are around 2 to 4 dollars.

There's a small number of higher tips (outliers), and the distribution is also right-skewed.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset("tips")

# Create a facet grid
g = sns.FacetGrid(data, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```

```python
# Load dataset
iris = sns.load_dataset("iris")

# Set up the grid
g = sns.PairGrid(iris, hue="species")

# Customize each section of the matrix
g.map_upper(sns.scatterplot)                 # Top triangle: scatter plots
g.map_lower(sns.kdeplot, cmap="Blues_d")     # Bottom triangle: KDE contours
g.map_diag(sns.histplot, kde=True)           # Diagonal: histograms with KDE

g.add_legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
tips = sns.load_dataset('tips')

# 1. Set context for publication
sns.set_context("paper")  # 'notebook', 'talk', 'poster' also available

# 2. Set whitegrid background for clarity
sns.set_style("whitegrid")

# 3. Set custom figure size and scale
sns.set(rc={"figure.figsize": (8, 5)})

# 4. Use a soft pastel color palette
sns.set_palette("pastel")

# 5. Create the plot
plot = sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="sex",                   # Use color by gender
    markers=["o", "s"],          # Custom marker shapes
    height=5, aspect=1.2,        # Size and proportion
    scatter_kws={"s": 50, "alpha": 0.7},   # Dot size and transparency
    line_kws={"lw": 2}                      # Regression line width
)

# 6. Title and axis labels
plt.title("Tip vs. Total Bill by Gender", fontsize=14, weight='bold')
plt.xlabel("Total Bill ($)", fontsize=12)
plt.ylabel("Tip ($)", fontsize=12)

# 7. Despine for cleaner edges
sns.despine()

#
plt.show()
```
