```python
import matplotlib.pyplot as plt

# Basic histogram using Matplotlib
plt.hist(data['total_bill'], edgecolor='black')  # default bins=10
plt.title('Histogram of Total Bill (Matplotlib)')
plt.xlabel('Total Bill')
plt.ylabel('Count')
#plt.grid(True)
plt.show()

# Customized histogram using Matplotlib
plt.hist(data['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Bill with Custom Bins and Color')
plt.xlabel('Total Bill')
plt.ylabel('Count')
#plt.grid(True)
plt.show()
```

Kernel Density Estimation (KDE) is a way to estimate the probability density function (PDF) of a continuous variable. In simpler terms:

It shows you where the data is concentrated — like a smoothed-out version of a histogram.

Instead of showing counts in bins (like a histogram), KDE draws a smooth curve that represents the shape of the data distribution.

```python
sns.kdeplot(data['total_bill'], fill=True)
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], bw_adjust=0.5, color='red', fill=True) # bw_Adjust determines how smooth the line will be (Default set to 1)
plt.title('Customized KDE of Total Bill')
plt.show()
```

```python
sns.displot(data['total_bill'], color='green', kde=True)
plt.title('Univariate Distribution of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter')
plt.title('Bivariate Distribution of Total Bill and Tips', pad=70)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
titanic = sns.load_dataset('titanic')

# Creating a bar plot
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='class', y='survived', hue='sex', data=titanic, palette='Set2', errorbar='sd')
plt.title('Survival Rate by Class and Gender', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# Creating a box plot
sns.boxplot(x='age', y='class', data=titanic, orient='h')
plt.title('Age Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='age', y='class', data=titanic, palette='Pastel1', width=0.6, orient='h')
sns.swarmplot(x='age', y='class', data=titanic, size=3, color='k', alpha=0.6, orient='h')
plt.title('Age Distribution by Class with Individual Data Points', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Passenger Class')
plt.tight_layout()
plt.show()
```

```python
# Creating a violin plot
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True,
               palette='coolwarm', inner='quartile', bw_method=0.8)
plt.title('Age Distribution by Class and Gender', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
data = sns.load_dataset('tips')

# Create scatter plot with regression line
sns.lmplot(x='total_bill', y='tip', data=data)

# Add labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Create a joint plot with regression line
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')

# Add a title
plt.subplots_adjust(top=0.9)
plt.suptitle('Joint Plot of Total Bill and Tip')
plt.show()
```

🔍 What This Plot Shows:
🟦 Main (Center) Plot: Scatter with Regression Line
Each point = one customer’s total_bill and tip.

The regression line shows the general positive relationship:
💬 As the total bill increases, tips also tend to increase.

The shaded area around the line is the confidence interval (default is 95%).

📈 Top Plot: Histogram + KDE of total_bill
Most bills are clustered between 10 and 20 dollars.

The distribution is right-skewed — fewer high bills, with a long tail on the right.

📊 Right Plot: Histogram + KDE of tip
Most tips are around 2 to 4 dollars.

There's a small number of higher tips (outliers), and the distribution is also right-skewed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------------
# 1. Load and Inspect Data
# -------------------------------------
url = "https://raw.githubusercontent.com/intern2grow/sales-data-analysis/main/sales_data.csv"
sales_data = pd.read_csv(url)

# Preview and inspect
print(sales_data.info())
print(sales_data.head())

# -------------------------------------
# 2. Data Preprocessing
# -------------------------------------
# Convert 'date' column to datetime
sales_data['date'] = pd.to_datetime(sales_data['date'])

# Check for and handle missing values
print("\nMissing values:\n", sales_data.isnull().sum())

# Drop or fill missing values (for example)
sales_data.dropna(inplace=True)

# -------------------------------------
# 3. Sales Trends Over Time
# -------------------------------------
# Aggregate revenue by date
daily_revenue = sales_data.groupby('date')['revenue'].sum().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=daily_revenue, x='date', y='revenue')
plt.title("Total Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -------------------------------------
# 4. Product Category Comparison
# -------------------------------------
category_revenue = sales_data.groupby('category')['revenue'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_revenue, x='category', y='revenue', palette='Set2')
plt.title("Revenue by Product Category")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# -------------------------------------
# 5. Relationship Between Price and Quantity Sold
# -------------------------------------
plt.figure(figsize=(8, 5))
sns.scatterplot(data=sales_data, x='price', y='quantity', hue='category', alpha=0.7)
plt.title("Price vs Quantity Sold")
plt.tight_layout()
plt.show()

# -------------------------------------
# 6. Product Performance Analysis (Pair Plot)
# -------------------------------------
sns.pairplot(sales_data[['price', 'quantity', 'revenue']], corner=True)
plt.suptitle("Pairwise Product Performance Metrics", y=1.02)
plt.show()

# -------------------------------------
# 7. Aggregated Category Insights
# -------------------------------------
category_avg_price = sales_data.groupby('category')['price'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_avg_price, x='category', y='price', palette='pastel')
plt.title("Average Price by Category")
plt.ylabel("Average Price")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
