```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Exercise 2: Creating a Line Plot for Temperature Variation

```python
# Create a simple line plot using Matplotlib that represents the temperature variations over a week.
temps = [72, 74, 76, 80, 82, 78, 75]
days = ['Sun', 'Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat']

plt.figure()
plt.plot(days, temps)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.show()
```

# Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
# Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store
sales = [5000, 5500, 6200, 7000, 7500, 8000, 8500, 8000, 10000, 9250, 8000, 7750]
month = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

plt.bar(month, sales)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

# Exercise 4: Data Visualization

```python
# Loading Dataset
url = 'https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%205%20-%20Data%20Processing/W5D1%20-%20Data%20Visualization,%20Matplot/sales_data.zip'
sales_data = pd.read_csv(url)
sales_data.head()
```

```python
# Calculate Total Quantity of Products Sold
total_quantity = int(sales_data['quantity'].sum())
print("Total Products Sold:", total_quantity)
```

```python
# Identify the category that had the highest revenue and calculate how much revenue it generated
revenues = sales_data.groupby('category')['revenue'].sum()
revenues.sort_values(ascending=False)
```

```python
# Determine the average revenue per sale in the dataset
avg_sale_revenue = round(sales_data['revenue'].mean(), 3)
print(avg_sale_revenue)
```

```python
# Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4)
sales_data['date'] = pd.to_datetime(sales_data['date'])

quarterly_sales = sales_data.groupby(pd.Grouper(key='date', freq='QE'))['revenue'].sum()

quarterly_sales = list(quarterly_sales)
quarter = ['Q1', 'Q2', 'Q3', 'Q4']

plt.bar(quarter, quarterly_sales)
plt.title("Quarterly Revenue - 2022")
plt.xlabel('Quarter')
plt.ylabel('Revenu')
plt.show()
```
