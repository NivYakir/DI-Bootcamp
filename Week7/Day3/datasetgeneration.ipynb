```python
import numpy as np
import pandas as pd
from faker import Faker
import faker_commerce
fake = Faker()
fake.add_provider(faker_commerce.Provider)
```

```python
# generating synthetic name and ages
data = [(fake.name(), np.random.randint(18, 70)) for _ in range(100)]

# converting to dataframe
df = pd.DataFrame(data, columns=['Name', 'Age'])
df.head()
```

```python
data = [(fake.ecommerce_name(), fake.pricetag(), fake.date_this_year(), np.random.randint(18, 70)) for _ in range(100)]
fake_data = pd.DataFrame(data, columns=['Product Name', 'Sale Amount', 'Date of Sale', 'Customer Age'])
fake_data.head(100)
```
